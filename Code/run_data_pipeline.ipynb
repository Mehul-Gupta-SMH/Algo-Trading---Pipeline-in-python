## Sample Run

```python
url = "https://alphastreet.com/india/earnings-call-transcripts/"

params_bs4_filter = {}
params_bs4_filter['name'] = 'a'
params_bs4_filter['href'] = True
params_bs4_filter['attrs'] = {'rel':'bookmark'}
params_bs4_filter['recursive'] = True

params_iter = {}
params_iter['scroll_wait_time'] = 5.0
params_iter['iter_threshold'] = 1
params_iter['silent_mode'] = True

params_debug = {}
params_debug['debug_mode'] = False
params_debug['ticker'] = 'KRBL'
```

```python
extracted_url_df , corp_ppts_df , analyst_ppts_df , disclosures_df , qa_df = get_transcripts_data_wrapper(url,params_iter,params_bs4_filter,params_debug)
```
